```python
# üìå Slot Machine Simulation

import random

# Simulated version (replace with original if given by environment)
def play_slot_machine():
    outcome = random.choices(
        population=[0, 1, 5, 10, 100],
        weights=[0.95, 0.03, 0.01, 0.009, 0.001],
        k=1
    )[0]
    return outcome

# üßÆ Monte Carlo Estimation
def estimate_average_slot_payout(n_runs):
    total_profit = 0
    for _ in range(n_runs):
        payout = play_slot_machine()
        net = payout - 1  # subtract cost to play
        total_profit += net
    return total_profit / n_runs

# ‚ñ∂Ô∏è Run Simulation
print("Estimated payout per play:", estimate_average_slot_payout(100000))
```
